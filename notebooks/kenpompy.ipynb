```python
from kenpompy.utils import login
import pandas as pd


# Returns an authenticated browser that can then be used to scrape pages that require authorization.
browser = login(EMAIL, PW)
```

```python
SNAPSHOT_DATE = "20260115"  # YYYYMMDD for 2026-01-15
SEASON = "2026"

url = f"https://kenpom.com/index.php?y={SEASON}&d={SNAPSHOT_DATE}"
resp = browser.get(url)

# Basic sanity checks
resp.raise_for_status()
html = resp.text
print("Fetched:", url)
print("HTML length:", len(html))

# Parse all tables; the Pomeroy ratings table is typically the first one.
tables = pd.read_html(html)
ratings_df = tables[0]

ratings_df.head()
```

```python
# Optional: clean column names a bit (KenPom tables can have repeated headers)
ratings_df.columns = [str(c).strip() for c in ratings_df.columns]

out_csv = "kenpom_ratings_2026-01-15.csv"
ratings_df.to_csv(out_csv, index=False)

print("Saved:", out_csv)
print("Rows:", len(ratings_df))
```
